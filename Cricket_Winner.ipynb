```python
import pandas as pd
train_data = pd.read_csv('Trainmatches.csv')
test_data = pd.read_csv('Testmatches.csv')
```

```python
train_data.head()
```

```python
features = ["team1", "team2", "toss_winner", "toss_decision", "venue", "dl_applied"]
target = "winner"
```

```python
test_data.head()
```

```python
train_data.isnull().sum()
```

```python
train_data.columns
```

```python
train_data = pd.get_dummies(train_data, columns=features)
test_data = pd.get_dummies(test_data, columns=features)
```

```python
test_data.rename(columns = {"result" : "winner"}, inplace=True)
```

```python
print("Train Dataset : ",train_data.columns)
print("Test Dataset : ",test_data.columns)
```

```python
missing_cols = set(train_data.columns) - set(test_data.columns)
for col in missing_cols:
    test_data[col] = 0
test_data = test_data[train_data.columns]
```

```python
drop_columns = ["id", "season", "city", "result", "win_by_runs", "win_by_wickets", "player_of_match"]
train_data = train_data.drop(columns=drop_columns, errors="ignore")
test_data = test_data.drop(columns=drop_columns, errors="ignore")
```

```python
train_data = train_data.dropna()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data.drop(columns=target), train_data[target], test_size=0.2, random_state=42)
```

```python
y_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
predictions = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(predictions, y_test)
accuracy
```

```python
train_pred = rf_classifier.predict(X_train)
train_accuracy = accuracy_score(train_pred, y_train)
train_accuracy
```

```python
import joblib
joblib.dump(rf_classifier, 'rf_classifier.pkl')
```
